```python
import pandas as pd
```

```python
cols=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm','species']
```

```python
df= pd.read_csv('iris.csv',sep=",",names=cols)
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
setosa = df[df["species"]=="Iris-setosa"]
versicolor = df[df["species"]=="Iris-versicolor"]
virginica = df[df["species"]=="Iris-virginica"]
```

```python
setosa.head()
versicolor.head()
virginica.head()
```

```python
versicolor.head()
```

```python
merged_data= pd.concat([setosa,versicolor])
```

```python
merged_data.tail()
```

```python
sorted_data= df.sort_values(by="petal length in cm",ascending=False)
```

```python
transpose_data=df.transpose()
```

```python
transpose_data.head()
```

```python
melted_data=pd.melt(df, id_vars=['species'],value_vars=['sepal length in cm', 'sepal width in cm', 'petal length in cm','petal width in cm'],
                      var_name='metric',
                      value_name='Value')
```

```python
melted_data.head()
```

```python
pivot_data = melted_data.pivot_table(index='species',columns='metric',values='Value', aggfunc='mean')

pivot_data
```
