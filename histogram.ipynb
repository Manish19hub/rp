```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("Cleavland.csv")

# a. Histogram
# Objective: To show distribution of Age

plt.hist(df.iloc[:,0], bins=10)

plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

# b. Dot Plot
# Objective: To display individual Cholesterol values

plt.plot(df.iloc[:,4], 'o')

plt.title("Dot Plot of Cholesterol")
plt.xlabel("Index")
plt.ylabel("Cholesterol")

plt.show()

# c. Bar Plot
# Objective: To compare Chest Pain Types

cp = df.iloc[:,2].value_counts()

plt.bar(cp.index, cp.values)

plt.title("Bar Plot of Chest Pain Type")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")

plt.show()

# d. Line Chart
# Objective: To show relationship between Age and Maximum Heart Rate

plt.plot(df.iloc[:,0], df.iloc[:,7])

plt.title("Line Chart of Age vs Thalach")
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate")

plt.show()

# e. Histogram and Scatter Plot with Box Plot
# Objective: To detect spread, outliers and relationship of data

plt.figure(figsize=(12,5))

# Histogram
plt.subplot(1,3,1)

plt.hist(df.iloc[:,4], bins=10)

plt.title("Histogram")

# Scatter Plot
plt.subplot(1,3,2)

plt.scatter(df.iloc[:,0], df.iloc[:,4])

plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

# Box Plot
plt.subplot(1,3,3)

plt.boxplot(df.iloc[:,4])

plt.title("Box Plot")

plt.show()
```

```python
# b. Pie Chart
# Objective: To show percentage distribution of Chest Pain Types

cp = df.iloc[:,2].value_counts()

plt.pie(cp.values, labels=cp.index, autopct='%1.1f%%')

plt.title("Pie Chart of Chest Pain Types")

plt.show()

# c. Box Plot
# Objective: To detect outliers in Cholesterol values

plt.boxplot(df.iloc[:,4])

plt.title("Box Plot of Cholesterol")

plt.show()

# d. Scatter Plot
# Objective: To show relationship between Age and Cholesterol

plt.scatter(df.iloc[:,0], df.iloc[:,4])

plt.title("Scatter Plot of Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

plt.show()

# e. Add Boxplot to Scatterplot
# Objective: To compare data spread and relationship together

plt.figure(figsize=(10,5))

# Scatter Plot
plt.subplot(1,2,1)

plt.scatter(df.iloc[:,0], df.iloc[:,4])

plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Cholesterol")

# Box Plot
plt.subplot(1,2,2)

plt.boxplot(df.iloc[:,4])

plt.title("Box Plot")

plt.show()
```
