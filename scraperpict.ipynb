```python
#pip install requests beautifulsoup4 pandas

import requests
from bs4 import BeautifulSoup
import pandas as pd

# Website URL
url = "https://www.shiksha.com/college/pune-institute-of-computer-technology-pict-pune-28383/reviews"

# Send Request
response = requests.get(url)

# Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Lists
names = []
ratings = []
reviews = []
tags = []

# Find review blocks
review_blocks = soup.find_all('div', class_='reviewTuple')

for block in review_blocks:

    # Customer Name
    try:
        name = block.find('span', class_='name').text.strip()
    except:
        name = "Anonymous"

    # Rating
    try:
        rating = block.find('span', class_='ratingNumber').text.strip()
    except:
        rating = "NA"

    # Review
    try:
        review = block.find('p').text.strip()
    except:
        review = "No Review"

    # Tag
    try:
        tag = block.find('a').text.strip()
    except:
        tag = "General"

    names.append(name)
    ratings.append(rating)
    reviews.append(review)
    tags.append(tag)

# Create DataFrame
df = pd.DataFrame({
    'Customer Name': names,
    'Rating': ratings,
    'Review': reviews,
    'Tag': tags
})

# Display Data
print(df)

# Save to CSV
df.to_csv("PICT_reviews.csv", index=False)

print("Scraping Completed")
```
